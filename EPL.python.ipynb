```python
#6 I saved a copy of the dataset in a new variable so as to avoid losing important information from the original dataset
epl_data = df.copy()
```

```python
#7 I focused on Chelsea Fc, since it is my favorite club, so i ran the following line of code to extract all data pertaining to chelsea from the home and away team column
chelsea_epl_data = epl_data[(epl_data["HomeTeam"] == "Chelsea") |
                           (epl_data["AwayTeam"] == "Chelsea")]
```

```python
#9 For better understanding of the data, I chose to work with more recent years, so i ran the following line of code to extract recent years from the seasons columns
chelsea_data = chelsea_epl_data[(chelsea_epl_data["Season"] >= "2015-16") &
                               (chelsea_epl_data["Season"] <= "2017-18")]
```

```python
#10 i called on the dataset variable to check if i worked correctly
chelsea_data.head()
```

```python
#11 From the dataset, i found out that the Div column does not contribute any meangiful knowledge 
chelsea_data = chelsea_data.drop("Div", axis = 1)
```

```python
chelsea_data.head()
```

```python
# 24 You will notice that the 'Half time home goals' and 'half time away goals' columns have floats(decimal points) as values so 
chelsea_data.info()
```

```python
#25 You will notice that their dtype is in float64, which will hinder editing of the half time result columns, so i converted them to int
chelsea_data["Half Time Home Goals"] = chelsea_data["Half Time Home Goals"].astype(int)
chelsea_data["Half Time Away Goals"] = chelsea_data["Half Time Away Goals"].astype(int)
```

```python
#25 To confim
chelsea_data.info()
```

```python
#26 Now, to edit the half time result to display the proper scores
chelsea_data["Half Time Result"] = chelsea_data["Half Time Home Goals"].astype(str) +"-"+ chelsea_data["Half Time Away Goals"].astype(str)
```

```python
#27 on calling the chelsea data, i noticed the date column is not in the pd datetime
chelsea_data.head()
```

```python
#28 so i converted it
chelsea_data['Date'] = pd.to_datetime(chelsea_data['Date'])
```

```python
# i called on the data to see if i worked correctly
chelsea_data.head()
```

```python
# i moved on to seperating the home team data from the away team with the following line of code
chelsea_home_data = chelsea_data[(chelsea_data["HomeTeam"] == "Chelsea")]
chelsea_away_data = chelsea_data[(chelsea_data["AwayTeam"]) == "Chelsea"]
```

```python
# Now that i have a clean and processed dataset, I will save them into a csv file
chelsea_data.to_csv('Chelsea-data.csv')
chelsea_home_data.to_csv('Chelsea-home-data.csv')
chelsea_away_data.to_csv('Chelsea-away-data.csv')
```
